```python
import pandas as pd

pd.read_csv("datasets/basketball.csv")
```

usecols - take the given column names

```python
pd.read_csv("datasets/basketball.csv", usecols=["Player"])
```

squeeze - converts a single column dataframe into series

```python
player_names = pd.read_csv("datasets/basketball.csv", usecols=["Player"], squeeze=True)
```

```python
player_names
```

```python
points = pd.read_csv("datasets/basketball.csv", usecols=["PTS"], squeeze=True)
```
