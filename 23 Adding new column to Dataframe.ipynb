```python
dataset = pd.read_csv("datasets/netflix.csv")
```

```python
dataset.head(3)
```

**Using <code>insert()</code> method to add new column:**

```python
dataset.insert(3, column="genre", value="Drama")
```

```python
dataset.head(3)
```

```python
dataset.insert(10, column="budget", value=100)
```

```python
dataset.head(3)
```

```python
dataset["collections"] = dataset["budget"] + 50
```

```python
dataset.head(3)
```
