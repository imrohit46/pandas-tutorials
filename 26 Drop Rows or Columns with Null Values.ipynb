# Drop Rows or Columns With Null Values
By:<a href='https://www.youtube.com/wonkyCode'>WonkyCode</a>

```python
import pandas as pd
import numpy as np
dataset = {
    "name" : ["Rohit", "Harsha", "Saran", "Abhijith", np.nan],
    "age" : [25, 18, np.nan, np.nan, np.nan],
    "height": [5.6, 5.8, 6.2, np.nan, np.nan],
    "job": [True, np.nan, np.nan, np.nan, np.nan]
}
data = pd.DataFrame(dataset)
```

```python
data.head()
```

* You can drop rows containing null values using <code>dropna()</code> method

```python
data.dropna()
```

```python
data
```

```python
data.dropna(how="all")
```

```python
data
```

```python
data.dropna(how="all", inplace=True)
```

```python
data
```

```python
dic = {
    "name" : ["Rohit", "Harsha", "Saran", "Abhijith", np.nan],
    "age" : [25, 18, np.nan, np.nan, np.nan],
    "height": [5.6, 5.8, 6.2, np.nan, np.nan],
    "job": [True, np.nan, np.nan, np.nan, np.nan]
}
data = pd.DataFrame(dic)
```

```python
data
```

```python
data.dropna(thresh=2) # requires that many non-null values in a row
```

```python
data.dropna(thresh=3)
```

```python
data
```

```python
data.dropna(how="all", inplace=True)
```

```python
data
```

```python
data.dropna(axis="columns")
```

```python
data
```

```python
data.dropna(subset=["height"])
```

```python
data
```

```python
data.dropna(subset=["age", "job"])
```
