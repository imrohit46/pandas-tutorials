```python
import pandas as pd

data = pd.read_csv("datasets/basketball.csv").dropna(how="all")
```

```python
data.dtypes
```

```python
data.info()
```

```python
data['birth_year'] = data['birth_year'].astype('int')
```

```python
data['birth_year'].fillna(0, inplace=True)
```

```python
data['birth_year'] = data['birth_year'].astype('int')
```

```python
data.head(3)
```

```python
data["Team"].nunique()
```

```python
data['Team'].fillna("None", inplace=True)
```

```python
data['Team'] = data['Team'].astype('category')
```

```python
data.head(3)
```

```python
data['birth_month'] = data['birth_month'].astype('category')
```
