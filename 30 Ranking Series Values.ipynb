Pandas <code>Dataframe.rank()</code> method returns a rank of every respective index of a series passed. The rank is returned on the basis of position after sorting.

By default, equal values are assigned a rank that is the average of the ranks of those values.

```python
import pandas as pd

data = pd.read_csv("datasets/basketball.csv").dropna(how="all")
```

```python
data.head()
```

```python
data["PTS"] = data["PTS"].fillna(0).astype("int")
```

```python
data.head()
```

```python
data["PTS"].max()
```

```python
data["PTS"].min()
```

```python
data["PTS"].rank()
```

```python
data["PTS"].rank(ascending=False)
```

```python
data["PTS"].rank(ascending=False).astype("int")
```

```python
data["Rank by PTS"] = data["PTS"].rank(ascending=False).astype("int")
```
